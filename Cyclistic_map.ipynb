```python
import pandas as pd
import glob

files = glob.glob('*.csv') # Make sure this is in the folder with your monthly data

cols_to_use = ['start_station_name', 'start_lat', 'start_lng']

map_lookup = pd.concat(
    (pd.read_csv(f, usecols=cols_to_use) for f in files), 
    ignore_index=True
).drop_duplicates(subset=['start_station_name'])

map_lookup.to_csv('chicago_station_locations.csv', index=False)

print("Success! 'chicago_station_locations.csv' is ready for Tableau.")
```
